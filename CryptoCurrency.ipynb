```python
df.drop(['Date','Name','Symbol','Volume','Marketcap','SNo'],axis=1,inplace=True)
```

```python
def create_features(df):
    df['dayofweek']=df.index.day_of_week
    df['month']=df.index.month
    df['dayofyear']=df.index.dayofyear

    return df
```

```python
df= create_features(df)
```

### model Training

```python
features=['High','Low','Open','dayofweek','month','dayofyear']
target='Close'
```

```python
x_train=df[features]
y_train=df[target]

x_test = new_df[features]
y_test=new_df[target]
```

```python
x_train
```

```python
import xgboost as xgb
```

```python
xgbr = xgb.XGBRFRegressor()
```

```python
xgbr.fit(x_train,y_train)
```

```python
#traing Accuracy
xgbr.score(x_train,y_train)
```

```python
#testing Accuracy
xgbr.score(x_test,y_test)
```

```python
xgbr=xgb.XGBRegressor(colsample_bytree=0.7,
                      learning_rate=0.01,
                      max_depth=3,
                      n_estimators=500,
                      objective='reg:squarederror')
```

```python
xgbr.fit(x_train,y_train)
```

```python
#testing accuracy
xgbr.score(x_test,y_test)
```

```python
extended_df=df._append(new_df)
extended_df
```

```python
#making a feature importance dataframe and plot  it to visualize the impact of each featureon model
fi=pd.DataFrame(data=xgbr.feature_importances_,
                index=xgbr.get_booster().feature_names,
                columns=['importance'])
fi.sort_values('importance').plot(kind='barh',title='Feature Importance')
plt.show()
```

```python
new_df['predictions']=xgbr.predict(x_test)
extended_df=extended_df.merge(new_df[['predictions']],how='left',left_index=True,right_index=True)
```

```python
ax=extended_df['Close'].loc[extended_df.index>'01-01-2021'].plot(figsize=(15,5))
extended_df['predictions'].plot(ax=ax,style='.')
plt.legend(['Actual Values','Predictions'])
ax.set_title('Entire dataaset against predictions')
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(x_train, y_train)
```

```python
rf_predictions = rf.predict(x_test)
rf_mse = mean_squared_error(y_test, rf_predictions)
rf_r2 = r2_score(y_test, rf_predictions)
```

```python
print(f"Random Forest MSE: {rf_mse}")
print(f"Random Forest R²: {rf_r2}")
```

```python
xgb_predictions = xgbr.predict(x_test)
xgb_mse = mean_squared_error(y_test, xgb_predictions)
xgb_r2 = r2_score(y_test, xgb_predictions)
```

```python
print(f"Random Forest MSE: {rf_mse}")
print(f"Random Forest R²: {rf_r2}")
print(f"XGBoost MSE: {xgb_mse}")
print(f"XGBoost R²: {xgb_r2}")
```
